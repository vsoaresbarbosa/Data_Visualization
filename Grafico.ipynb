```python
#importando as bibliotecas

import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

```python
#Grafico com a Matplotlib - plotagem de função matemática (integração com numpy)
#Criar matriz com função linspace da numpy

t = np.linspace(0.0, 2.0, 100)

#Criar outras matrizes com diferentes funções trigonometricas 

f1 = np.sin(np.pi * t)
f2 = np.sin(2 * np.pi * t)

plt.figure(figsize=(10,8))

plt.plot(t, f1, linewidth=4.0, label='Seno de PI * T')
plt.plot(t, f2, linewidth=4.0, label='2 Seno de PI * T')

plt.legend()

plt.title('Senoides')

plt.grid(b=True, color='gray', linestyle='--', linewidth=0.5);
```

```python
#Grafico Violin
credit = pd.read_excel('/content/ClientesRenomeados.xlsx')
```

```python
#Objetivo: analisar a varição do valor de credito por estado civil

fig = px.violin(credit, x='estado_civil', y='Valor_credito', color='SEX', box=True,title='Valor de Crédito por Estado Civil')

fig.show()
```

```python
#GRafico com a Plotly - Scatter - Dispersão

#Sintese de situçaõ do cliente em stembro 2005
#express- px

fig = px.scatter(credit, x='status_pag_set_2005', y='valor_pag_set_2005', color='estado_civil', size='Valor_credito', title='Síntese da situação dos clientes em Setembro 2005')
fig.show()


#Salvar Grafico iterativo

fig.write_html('sintese_cliente_set_2005.html')
```
